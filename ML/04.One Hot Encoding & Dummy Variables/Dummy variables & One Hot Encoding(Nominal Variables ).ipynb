## One Hot Encoding with Pandas

```python
import pandas as pd
```

```python
tips_df = pd.read_csv(r'tips.csv')
tips_df.head()
```

```python
dummy_df1 = pd.get_dummies(tips_df)
dummy_df1.head()
```

```python
dummy_df2 = pd.get_dummies(tips_df, drop_first=True)
dummy_df2.head()
```

## One Hot Encoding with Sciki-learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc1 = OneHotEncoder(sparse=False)
```

```python
enc_arr1 = enc1.fit_transform(tips_df[['sex','smoker','day','time']])
enc_arr1
```

```python
dummy_df1.keys()
```

```python
enc_df1 = pd.DataFrame(enc_arr1, columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
enc_df1.head()
```

```python
enc2 = OneHotEncoder(sparse=False, drop='first')
```

```python
enc_arr2 = enc2.fit_transform(tips_df[['sex','smoker','day','time']])
enc_arr2
```

```python
enc_df2 = pd.DataFrame(enc_arr2, columns=['sex_Male', 'smoker_Yes', 'day_Sat',
                     'day_Sun', 'day_Thur', 'time_Dinner'])
enc_df2.head()
```
