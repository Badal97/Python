```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.violinplot(x = 'total_bill',data = tips)
plt.show()
```

```python
sns.violinplot(y='tip',data=tips)
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data = tips)
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data = tips,hue='day')
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data = tips,hue='sex')
plt.show()
```

```python
sns.violinplot(x = 'size', y = 'total_bill',data = tips,)
plt.show()
```

```python
sns.violinplot(x = 'size', y = 'total_bill',data = tips,hue='sex')
plt.show()
```

```python
sns.violinplot(x = 'size', y = 'total_bill',data = tips,hue='day')
plt.show()
```

```python
sns.violinplot(x = 'sex', y = 'total_bill',data = tips,)
plt.show()
```

```python
sns.violinplot(x = 'sex', y = 'total_bill',data = tips,hue='day')
plt.show()
```

```python
sns.violinplot(x = 'sex', y = 'total_bill',data = tips,hue='sex')
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data = tips,hue='sex',palette='husl')
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data = tips,hue='smoker',palette='husl')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True,palette='coolwarm')
plt.show()
```

```python
order = ['Sun','Fri','Thur','Sat']
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm',order=order)
plt.show()
```

```python
order = ['Sun','Fri','Thur','Sat']
sns.violinplot(x='day',y='total_bill',data=tips,palette='RdBu',order=order,hue='smoker',
               split=True)
plt.show()
```

```python
order = ['Sun','Fri','Thur','Sat']
sns.violinplot(x='day',y='total_bill',data=tips,palette='RdBu',order=order,hue='smoker',
               split=True,inner='quartile')
plt.show()
```

```python
order = ['Sun','Fri','Thur','Sat']
sns.violinplot(x='day',y='total_bill',data=tips,palette='RdBu',order=order,hue='smoker',
               split=True,inner='box')
plt.show()
```

```python
order = ['Sun','Fri','Thur','Sat']
sns.violinplot(x='day',y='total_bill',data=tips,palette='RdBu',order=order,hue='smoker',
               split=True,inner='point')
plt.show()
```

```python
order = ['Sun','Fri','Thur','Sat']
sns.violinplot(x='day',y='total_bill',data=tips,palette='RdBu',order=order,hue='smoker',
               split=True,inner='quartile',scale='count')
plt.show()
```
