```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ml_students_age = np.random.randint(18,45, (100))
py_students_age = np.random.randint(15,40, (100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
plt.hist(py_students_age)
plt.title("Python Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
bins= [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,color='m',rwidth=0.8, histtype = "bar",label="ML Student")
plt.title('ML students age Histogram')
plt.xlabel('Student age Category')
plt.ylabel('No Student age')
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
bins= [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,color='m',rwidth=0.8, histtype = "bar",label="ML Student",orientation='vertical')
plt.title('ML students age Histogram')
plt.xlabel('Student age Category')
plt.ylabel('No Student age')
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
bins= [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,color='m',rwidth=0.8, histtype = "bar",label="ML Student",orientation='horizontal')
plt.title('ML students age Histogram')
plt.xlabel('Student age Category')
plt.ylabel('No Student age')
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m", "y"], label = ["ML Student", "Py Student"])
plt.title("ML &amp; Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.savefig('Image/hist2.png')
plt.show()
```
