```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.FacetGrid(tip,row ='smoker',col='time')
plt.show()
```

```python
x = sns.FacetGrid(tip,row ='smoker',col='time')
x = x.map(plt.hist,'total_bill')
plt.show()
```

```python
x = sns.FacetGrid(tip,row ='smoker',col='time')
x = x.map(plt.hist,'total_bill',color='green')
plt.show()
```

```python
x = sns.FacetGrid(tip,row ='smoker',col='time')
x = x.map(plt.hist,'total_bill',bins=15,color='blue')
plt.show()
```

```python
x = sns.FacetGrid(tip,row='smoker',col='time')
x = x.map(plt.scatter,'total_bill','tip',color='green')
plt.show()
```

```python
x = sns.FacetGrid(tip,row='smoker',col='time')
x = x.map(sns.regplot,'total_bill','tip')
plt.show()
```

```python
x = sns.FacetGrid(tip,col='smoker',hue='time')
x = x.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
x = sns.FacetGrid(tip,col='time',hue='smoker')
x = x.map(plt.scatter,'total_bill','tip')
x = x.add_legend()
plt.show()
```

```python
x = sns.FacetGrid(tip,col='day')
x = x.map(sns.boxplot,'total_bill','time')
plt.show()
```

```python
x = sns.FacetGrid(tip,col='day',size=5)
x = x.map(sns.boxplot,'time','total_bill')
plt.show()
```

```python
x = sns.FacetGrid(tip,col='day',size=7,aspect=.2)
x = x.map(sns.boxplot,'time','total_bill')
plt.show()
```

```python
order = ['Sun','Thur','Fri','Sat']
x = sns.FacetGrid(tip,col='day',col_order = order,aspect = 0.4,size=7)
x = x.map(sns.boxplot,'time','total_bill',color='red')
plt.show()
```

```python
x = sns.FacetGrid(tip,col='day',hue='smoker',size=6,aspect=0.4,palette='cool')
x = x.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
x = sns.FacetGrid(tip,col='time',hue='smoker',size=5,palette='coolwarm')
x = x.map(plt.scatter,'total_bill','tip')
x = x.add_legend()
plt.show()
```

```python
x = sns.FacetGrid(tip,col='time',hue='smoker',size=5,palette='husl')
x = x.map(plt.scatter,'total_bill','tip')
x = x.add_legend()
plt.show()
```
