```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('googleplaystore.csv',nrows=1000)
df.head(2)
```

```python
df.shape
```

```python
df.columns
```

```python
x = df['Rating']
y = df['Reviews']
```

```python
plt.scatter(x,y)
plt.title('Google Play Store Apps Scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(11,7))
plt.scatter(x, y, c='r', marker='*',alpha=0.5,linewidths=8,s=100,edgecolors='g')
plt.title('Google Play Store Apps Scatter plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.grid()
plt.show()
```

```python
z = df["Installs"]
plt.scatter(x,y, c = "m", marker = "*", s = 100, alpha=0.5, linewidths=10,edgecolors="b" )
plt.scatter(z,x, c = "r", marker = "o", s = 100, alpha=0.5, linewidths=10,edgecolors="y" )

plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.show()
```
