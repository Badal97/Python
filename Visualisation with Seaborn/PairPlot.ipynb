```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.pairplot(tip)
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.pairplot(iris,hue='species',diag_kind='hist')
plt.show()
```

```python
sns.pairplot(iris,hue='species',palette='husl')
plt.show()
```

```python
sns.pairplot(iris,hue='species',diag_kind='hist',palette='RdBu')
plt.show()
```

```python
sns.pairplot(iris,hue='species',markers=['o','<','s'],palette='husl')
plt.show()
```

```python
sns.pairplot(iris,hue='species',vars=['sepal_length','sepal_width'],)
plt.show()
```

```python
sns.pairplot(tip,vars=['total_bill','tip'],)
plt.show()
```

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width'],)
plt.show()
```

```python
sns.pairplot(iris,height=5,vars=['sepal_length','sepal_width'],)
plt.show()
```

```python
sns.pairplot(iris,x_vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width'])
plt.show()
```

```python
sns.pairplot(iris,x_vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width'],
            hue='species',palette='husl')
plt.show()
```

```python
sns.pairplot(iris,diag_kind='kde')
plt.show()
```

```python
sns.pairplot(iris,diag_kind='kde',palette='husl',hue='species')
plt.show()
```

```python
sns.pairplot(iris,kind='reg')
plt.show()
```

```python
sns.pairplot(iris,diag_kind='kde',palette='husl',hue='species',kind='reg')
plt.show()
```
