## Strip Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.stripplot(x = tips['tip'],color='green',)
plt.show()
```

```python
sns.stripplot(x = tips['total_bill'],color='red')
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,jitter=False)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,jitter=True)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,jitter=0.3)
plt.show()
```

```python
sns.stripplot(y = 'day',x='total_bill',data = tips,jitter = 0.6)
plt.show()
```

```python
sns.stripplot(y = 'day',x='total_bill',data = tips,linewidth=1.2)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,hue='sex')
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,hue='sex',jitter=False,dodge=True)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,hue='smoker',jitter=False,dodge=True)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,hue='smoker',jitter=False,dodge=True,
                    palette='gist_ncar_r')
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,hue='sex',jitter=False,marker='D',size=12)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,hue='sex',jitter=False,marker='D',size=12,
             edgecolor='green',alpha=0.4)
plt.show()
```

```python
sns.stripplot(x = 'tip',y='day',data = tips,jitter = True,color='.4')
sns.boxplot(x = 'tip',y='day',data = tips)
plt.show()
```

```python
sns.stripplot(x = 'day',y='total_bill',data = tips,color='.2')
sns.violinplot(x = 'day',y='total_bill',data = tips)
plt.show()
```
