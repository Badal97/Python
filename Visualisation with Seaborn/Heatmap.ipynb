```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.rand(12,15)
```

```python
sns.heatmap(x)
plt.show()
```

```python
sns.heatmap(x,annot=True)
plt.show()
```

```python
sns.heatmap(x,vmin =0, vmax=3)
plt.show()
```

```python
sns.heatmap(x ,vmin =0,vmax = 0.5 )
plt.show()
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
flight.tail()
```

```python
flight.columns
```

```python
flights = flight.pivot('month','year','passengers')
sns.heatmap(flights)
plt.show()
```

```python
sns.heatmap(flights,annot=True)
plt.show()
```

```python
sns.heatmap(flights,annot = True,fmt = 'd')
plt.show()
```

```python
sns.heatmap(flights,annot = True,fmt = '0.1f')
plt.show()
```

```python
sns.heatmap(flights,linewidth=0.9)
plt.show()
```

```python
sns.heatmap(flights,annot = True,fmt = 'd',linewidth=0.9)
plt.show()
```

```python
sns.heatmap(flights,annot = True,fmt = 'd',linewidth=1,vmin=100,vmax=650)
plt.show()
```

```python
sns.heatmap(flights,cmap = 'RdBu')
plt.show()
```

```python
sns.heatmap(flights,cmap='RdBu',annot = True,fmt = 'd')
plt.show()
```

```python
sns.heatmap(flights,cmap='summer',annot = True,fmt = 'd')
plt.show()
```

```python
sns.heatmap(flights,cmap='winter_r',annot = True,fmt = 'd')
plt.show()
```

```python
sns.heatmap(flights,cmap='coolwarm',annot = True,fmt = 'd')
plt.show()
```

```python
sns.heatmap(flights,center = flights.loc['June',1954],annot= True,fmt='d')
plt.show()
```

```python
sns.heatmap(flights,center = flights.loc['March',1959],annot= True,fmt='d')
plt.show()
```

```python
sns.heatmap(flights,center = flights.loc['June',1954],annot= True,fmt='d',cbar = False)
plt.show()
```

```python
sns.heatmap(flights,cbar=False)
plt.show()
```

```python
sns.heatmap(flights , cbar = False, linewidth = 1.2)
plt.show()
```
