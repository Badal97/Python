```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.barplot(x = df['day'], y = df['total_bill'])
plt.show()
```

```python
sns.barplot(x = df['sex'], y = df['tip'])
plt.show()
```

```python
sns.barplot(x = df['day'], y = df['total_bill'],hue= df['sex'])
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = df)
plt.show()
```

```python
order= ['Sun','Sat','Thur','Fri']
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = df,order=order)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = df,hue_order=['Female','Male'],)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = df,estimator = np.max)
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = df,ci=50,n_boot=10,)
plt.show()
```

```python
sns.barplot(y ='day',x='total_bill',hue='sex',data=df,)
```

```python
sns.barplot(x ='total_bill', y = 'size', hue = 'sex', data = df,orient='h')
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,color='g')
plt.show()
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,palette='magma')
plt.show()
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,palette='magma',saturation=0.4)
plt.show()
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,errcolor='0.4',)
plt.show()
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,errwidth=12)
plt.show()
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,capsize=0.2)
plt.show()
```

```python
sns.barplot(x = 'size',y ='total_bill', hue = 'sex', data = df,palette='magma',dodge=False)
plt.show()
```

```python
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',data = df,**kwargs)
plt.show()
```

```python
sns.set()
ax = sns.barplot(x = 'day', y = 'total_bill',data = df, alpha =.9, linestyle = "-.", linewidth=3,
           edgecolor = "g")
 
ax.set(title = "Barplot of Tips DataFrame",xlabel = "Days",ylabel = "Total Bill")
plt.show()
```

```python
sns.set()
plt.figure(figsize = (16,9))
sns.barplot(x = 'day', y = 'total_bill', data = df, alpha =1, linestyle = "-.", linewidth = 3,
           edgecolor = "k")
 
plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
 
plt.savefig("Image/Barplot of Days and Total Bill")
plt.show()
```
