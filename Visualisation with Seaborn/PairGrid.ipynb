```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
sns.set()
```

```python
x = sns.PairGrid(iris)
x.map(plt.scatter)
plt.show()
```

```python
x = sns.PairGrid(iris)
x.map_diag(plt.hist)
plt.show()
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
plt.show()
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
plt.show()
```

```python
x = sns.PairGrid(iris,hue='species',palette='husl')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
plt.show()
```

```python
x = sns.PairGrid(iris,hue='species',palette='coolwarm')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
plt.show()
```

```python
x = sns.PairGrid(iris,hue='species',palette='autumn')
x = x.map_diag(plt.hist,histtype='step')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
plt.show()
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist,histtype='step',linewidth=4)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
plt.show()
```

```python
x = sns.PairGrid(iris,vars=['petal_length','petal_width'])
x = x.map(plt.scatter)
plt.show()
```

```python
x = sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
plt.show()
```

```python
x = sns.PairGrid(iris,vars=['petal_length','petal_width'],hue='species')
x = x.map_diag(plt.hist,edgecolor='white')
x = x.map_offdiag(plt.scatter,edgecolor='black')
x = x.add_legend()
plt.show()
```

```python
x =sns.PairGrid(iris,x_vars=['petal_length','petal_width'],
                y_vars=['sepal_length','sepal_width'])
x = x.map(plt.scatter)
plt.show()
```

```python
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue='species',palette='coolwarm')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue='species')
x = x.map(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','*']})
x = x.map(plt.scatter)
x = x.add_legend()
```
