## Box Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.set()
df = sns.load_dataset('tips')
df.head()
```

```python
sns.boxplot(x = 'size', data = df)
plt.show()
```

```python
sns.boxplot(x = 'total_bill', data = df)
plt.show()
```

```python
sns.boxplot(x = 'sex',y = 'total_bill', data = df)
plt.show()
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df)
plt.show()
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df,hue='sex')
plt.show()
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df,hue='sex',palette='CMRmap_r')
plt.show()
```

```python
sns.boxplot(x='day',y= 'total_bill',data=df,hue='time')
plt.show()
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df,order =['Sun','Fri','Thur','Sat'])
plt.show()
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data = df,order =['Sun','Fri','Thur','Sat'],
                     palette='CMRmap_r')
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data= df)
plt.show()
```

```python
sns.boxplot(x='sex',y='tip',data= df,order = ['Female','Male'])
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.boxplot(data =iris)
plt.show()
```

```python
sns.boxplot(data =iris,palette ='coolwarm')
plt.show()
```

```python
sns.boxplot(data =iris,orient='horizontal',palette='husl')
plt.show()
```

```python
sns.boxplot(data =iris,orient='v',palette='husl')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data = df,palette='husl')
sns.swarmplot(x='day',y='total_bill',data = df)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data = df,palette='husl')
sns.swarmplot(x='day',y='total_bill',data = df,color='black')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data = df,palette='husl')
sns.swarmplot(x='day',y='total_bill',data = df,color='0.35')
plt.show()
```
