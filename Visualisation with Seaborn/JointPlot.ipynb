```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
plt.show()
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tip,kind='reg')
plt.show()
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tip,kind='reg',color='green')
plt.show()
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tip,kind='reside')
plt.show()
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tip,kind='hex')
plt.show()
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=tip,kind='kde')
plt.show()
```

```python
sns.jointplot(x = 'sepal_length',y = 'sepal_width',data = iris,kind = 'kde')
plt.show()
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='total_bill',y='tip',data = tip, stat_func = spearmanr)
plt.show()
```

```python
sns.jointplot(x='size',y='total_bill',data = tip, stat_func = spearmanr)
plt.show()
```

```python
sns.jointplot(x='size',y='total_bill',data = tip, stat_func = spearmanr,color='green')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data = tip, ratio=1,size=6)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data = tip, ratio=3,size=6)
plt.show()
```
