## scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.scatterplot(x= 'age',y='fare',data= df)
plt.show()
```

```python
sns.scatterplot(x= 'age',y='fare',data= df,hue='sex')
plt.show()
```

```python
sns.scatterplot(x= 'age',y='fare',data= df,hue='sex',style='who')
plt.show()
```

```python
sns.scatterplot(x= 'age',y='fare',data= df,hue='sex',style='who',size='who')
plt.show()
```

```python
sns.scatterplot(x= 'age',y='fare',data= df,hue='sex',style='who',size='who',sizes=(100,300),)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x= 'age',y='fare',data= df,hue='sex',style='who',size='who',sizes=(100,300),)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x= 'who',y='fare',data= df,hue='alive',style='alive',size='who',sizes=(100,300),)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x= 'who',y='fare',data= df,hue='alive',style='alive',size='who',sizes=(100,300),
               palette='YlOrBr',alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x= 'age',y='fare',data= df,hue='sex')
sns.lineplot(x= 'age',y='fare',data= df,hue='sex')
sns.barplot(x= 'age',y='fare',data= df,hue='sex')
plt.savefig('Image/Combination')
plt.show()
```
