## Histogram / Distplot

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as norm
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.distplot(df.tip)
plt.show()
```

```python
sns.distplot(df['size'])
plt.show()
```

```python
sns.distplot(df.total_bill)
plt.show()
```

```python
sns.distplot(df['total_bill'],bins=50)
plt.show()
```

```python
sns.distplot(df['total_bill'],hist=False,rug=True)
plt.show()
```

```python
sns.distplot(df['total_bill'],kde=False)
plt.show()
```

```python
# sns.distplot(df['total_bill'],fit = norm, kde = False)
```

```python
sns.distplot(df["total_bill"],color="r",)
plt.show()
```

```python
sns.distplot(df['total_bill'],vertical=True)
plt.show()
```

```python
sns.distplot(df['total_bill'],norm_hist=True)
plt.show()
```

```python
sns.distplot(df['total_bill'],axlabel= 'Total_Bill')
```

```python
sns.distplot(df['total_bill'],label='Total_Bill')
plt.title('Histogram of Total Bill')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,9))
sns.set()
sns.distplot(df['total_bill'],label='Total Bill',color='r',)
plt.title('Histogram of Total Bill')
plt.legend()
plt.savefig('Image/Histogram1')
plt.show()
```

```python
df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55] 
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(df["total_bill"], bins = bins)
 
plt.xticks(bins)
plt.title("Histogram of Total Bill",fontsize=20) 
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()
 # hist keyword argument to change hist format
sns.distplot(df["total_bill"],
        hist_kws = {'color':'r', 'edgecolor':'y','linewidth':5, 'linestyle':'--', 'alpha':0.9}) 
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
# hist keyword argument to change hist format
sns.distplot(df["total_bill"],
        hist_kws ={'color':'r', 'edgecolor':'y','linewidth':5, 'linestyle':'--', 'alpha':0.9},
        kde_kws ={'color':'b','linewidth':10, 'linestyle':'--', 'alpha':0.9})   
```

```python
plt.figure(figsize=(12,7))
sns.set()
# hist keyword argument to change hist format
sns.distplot(df["total_bill"],
        hist_kws ={'color':'r', 'edgecolor':'y','linewidth':5, 'linestyle':'--', 'alpha':0.9},
        kde_kws ={'color':'b','linewidth':7, 'linestyle':'--', 'alpha':0.9},
        rug=True,
        rug_kws ={'color':'k', 'edgecolor':'m','linewidth':5, 'linestyle':'--', 'alpha':0.9},
            ) 
```

```python
plt.figure(figsize=(16,9))
sns.distplot(df["total_bill"], bins=9, label="total_bil")
sns.distplot(df["tip"], bins=9, label="tip")
sns.distplot(df["size"], bins=10, label = "size")
 
plt.legend()
plt.savefig('Image/Distpot')
plt.show()
```
